```python
import requests
import pandas as pd

# Pull club Elo ratings for dates just before each WC
dates = ["2006-06-01", "2010-06-01", "2014-06-01", "2018-06-01", "2022-11-01"]

all_elo = []
for date in dates:
    url = f"http://api.clubelo.com/{date}"
    df = pd.read_csv(url)
    df["snapshot_date"] = date
    all_elo.append(df)
    print(f"{date}: {len(df)} clubs")

club_elo = pd.concat(all_elo, ignore_index=True)
club_elo.to_csv(
    "/Users/banna/Desktop/player_analysis/data/Derivedclub_elo_ratings.csv", index=False
)
print(f"\nSaved: {club_elo.shape}")
print(club_elo.head())
```
